```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import edhec_risk_kit as erk
import pandas as pd
import numpy as np

ind_return=erk.get_ind_returns()
ind_size=erk.get_ind_size()
ind_nfirms=erk.get_ind_nfirms()
```

```python
ind_mktcap=ind_nfirms*ind_size
ind_mktcap.shape
```

```python
total_mktcap=ind_mktcap.sum(axis="columns")
```

```python
total_mktcap.shape
```

```python
total_mktcap.plot(figsize=(10,5))
```

```python
ind_capweight=ind_mktcap.divide(total_mktcap, axis=0)
```

```python
ind_capweight.head()
```

```python
ind_capweight.loc["1926"].sum(axis="columns")
```

```python
ind_capweight[["Fin","Steel"]].plot(figsize=(12,5))
```

```python
total_market_return=(ind_capweight*ind_return).sum(axis=1)
```

```python
total_market_return.head()
```

```python
total_market_index=erk.drawdown(total_market_return).Wealth
```

```python
total_market_index.plot(figsize=(10,5), title=("Total mkt return"))
```

```python
total_market_index["1980":].plot(figsize=(10,5))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(10,5))
```

```python
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets,periods_per_year=12)
tmi_tr36rets.plot(figsize=(12,6), label="Tr36 mo Returns",legend=True)
#total_market_return.plot(label="Returns", legend=True)
```

## Rolling correlation - along with Multiindexes and `.groupby`

```python
ts_corr=ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
ts_corr.index.names=['date','industry']
ts_corr.tail()
```

```python
ind_tr36corr=ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label="Tr36 Mo Rets", figsize=(12,6),legend=True)
ind_tr36corr.plot(label="Tr36 Mo Rets", legend=True,secondary_y=True)
```

```python
tmi_tr36rets.corr(ind_tr36corr)
```
